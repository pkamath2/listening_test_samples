```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
import soundfile
from IPython.display import Audio
librosa.__version__
```

```python
directory = 'data/trumpinet-gan/'
audio_data = []
selected_index = [7, 9, 11, 13, 15]
# for f_num in range(0, 20, 2):
ids = ''
for f_num in selected_index:
    ids = ids + '_' + str(f_num)
    file = 'test_pitch_sweep_76_{}_0.wav'.format(f_num)
#     d, _ = librosa.load(os.path.join(directory,file), sr=16000, offset=0.05, duration=0.5)
    d, _ = librosa.load(os.path.join(directory,file), sr=16000)
    audio_data = np.append(audio_data, d)
#     audio_data = np.append(audio_data, np.zeros(1600))

librosa.display.waveplot(audio_data, sr=_)
soundfile.write('test_pitch_sweep_76'+ids+'.wav', audio_data, samplerate=16000)
Audio(audio_data, rate=_)
```
